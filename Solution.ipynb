### Используем данные Росстат (октябрь 2019) средних зарплат в зависимости от образования для заполнения NaN в графе образование:

```python
rosstat_edu = {'Высшее': 57700,
              'Среднее профессиональное': 36900,
              'Среднее': 34800}
#Данных о незаконченном высшем, к сожалению нет, поэтому придется считать среднюю зарплату по соответствующим работникам из train
rosstat_edu['Незаконченное высшее'] = train[train['education_type'] == 'Незаконченное высшее']['salary'].mean()
print('Средняя зарплата работников с незаконченным высшим:', rosstat_edu['Незаконченное высшее'])
def transform1(df):
    for i in df[pd.isna(df['education_type'])].index:
        if (df.at[i, 'salary'] > rosstat_edu['Высшее']):
            df.at[i, 'education_type'] = 'Высшее'
        elif (df.at[i, 'salary'] > rosstat_edu['Незаконченное высшее']):
            df.at[i, 'education_type'] = 'Незаконченное высшее'
        elif (df.at[i, 'salary'] > rosstat_edu['Среднее профессиональное']):
            df.at[i, 'education_type'] = 'Среднее профессиональное'
        else:
            df.at[i, 'education_type'] = 'Среднее'
    return df

def transform1t(df):
    for i in df[pd.isna(df['education_type'])].index:
        if (df.at[i, 'salary_desired'] > rosstat_edu['Высшее']):
            df.at[i, 'education_type'] = 'Высшее'
        elif (df.at[i, 'salary_desired'] > rosstat_edu['Незаконченное высшее']):
            df.at[i, 'education_type'] = 'Незаконченное высшее'
        elif (df.at[i, 'salary_desired'] > rosstat_edu['Среднее профессиональное']):
            df.at[i, 'education_type'] = 'Среднее профессиональное'
        else:
            df.at[i, 'education_type'] = 'Среднее'
    return df


train = transform1(train)
test = transform1t(test)
```

#### Действительно, есть. Нужно их заменить.
#### Если возраст указан, заменим опыт на медианный опыт для группы +- 2 года
#### Если не указан, заменим на медианный опыт работы

```python
median_experience = train['experience'].median()
median_experience_groups = []
for i in range(8):
    median_experience_groups.append(train[(train['age'] - 18) // 5 == i]['experience'].median())
median_experience_groups.append(train[(train['age'] - 18) // 5 >= 8]['experience'].median())
def transform5(df):
    for i in df[(pd.isna(df['experience'])) | (df['experience'] > 50)].index:
        if (pd.isna(df.at[i, 'age']) or df.at[i, 'age'] >= 100 or df.at[i, 'age'] < 18):
            df.at[i, 'experience'] = median_experience
        else:
            if (df.at[i, 'age'] >= 58 and df.at[i, 'age'] < 100):
                df.at[i, 'experience'] = median_experience_groups[8]
            elif (df.at[i, 'age'] >= 18):
                df.at[i, 'experience'] = median_experience_groups[int((df.at[i, 'age'] - 18) // 5)]
    return df
                
train = transform5(train)
test = transform5(test)
```

#### Заменим NaN значения в возрасте на средний возраст по опыту работы, выбросы заменим на медианный возраст

```python
median_age = train['age'].median()
# группы по опыту работы: 0 лет,(0-2]лет, 2-5лет,5-10лет, 10-15лет, 15-20лет, 20-25лет, 25-35лет, больше 
median_group1 = train[train['experience'] == 0]['age'].median()
median_group2 = train[(train['experience'] > 0) & (train['experience'] <= 2)]['age'].median()
median_group3 = train[(train['experience'] > 2) & (train['experience'] <= 5)]['age'].median()
median_group4 = train[(train['experience'] > 5) & (train['experience'] <= 10)]['age'].median()
median_group5 = train[(train['experience'] > 10) & (train['experience'] <= 15)]['age'].median()
median_group6 = train[(train['experience'] > 15) & (train['experience'] <= 20)]['age'].median()
median_group7 = train[(train['experience'] > 20) & (train['experience'] <= 25)]['age'].median()
median_group8 = train[(train['experience'] > 25) & (train['experience'] <= 35)]['age'].median()
median_group9 = train[train['experience'] > 35]['age'].median()

def transform6(df):
    for i in df[(pd.isna(df['age'])) | (df['age'] < 18) | (df['age'] > 85)].index:
        if (df.at[i,'experience'] == 0):
            df.at[i, 'age'] = median_group1
        if ((df.at[i, 'experience'] > 0)|(df.at[i, 'experience'] <= 2)):
            df.at[i, 'age'] = median_group2
        if ((df.at[i, 'experience'] > 2)|(df.at[i, 'experience'] <= 5)):
            df.at[i, 'age'] = median_group3
        if ((df.at[i, 'experience'] > 5)|(df.at[i, 'experience'] <= 10)):
            df.at[i, 'age'] = median_group4
        if ((df.at[i, 'experience'] > 10)|(df.at[i, 'experience'] <= 15)):
            df.at[i, 'age'] = median_group5
        if ((df.at[i, 'experience'] > 15)|(df.at[i, 'experience'] <= 20)):
            df.at[i, 'age'] = median_group6
        if ((df.at[i, 'experience'] > 20)|(df.at[i, 'experience'] <= 25)):
            df.at[i, 'age'] = median_group7
        if ((df.at[i, 'experience'] > 25)|(df.at[i, 'experience'] <= 35)):
            df.at[i, 'age'] = median_group8
        if (df.at[i, 'experience'] > 35):
            df.at[i, 'age'] = median_group9
    return df

train = transform6(train)
test = transform6(test)
```

```python
def transform7(df):
    df['drive_licence_A'] = 0
    df['drive_licence_B'] = 0
    df['drive_licence_C'] = 0
    df['drive_licence_D'] = 0
    df['drive_licence_E'] = 0

    for i in df[pd.isna(df['drive_licences']) == False].index:
        for licence in df.at[i, 'drive_licences'][1:-1].split(', '):
            df.at[i, 'drive_licence_' + licence] = 1
    return df

train = transform7(train)
test = transform7(test)
```
